**1.** $\text{ }$ Find the eigenvalues and eigenfunctions of the problem:

> $\begin{align*}
 u'' +\lambda u &= 0\\ 
 u(1) -u(0) &= 0\\ 
 u'(1) -u'(0) &= 0 
\end{align*}$

**solution**

>$\begin{align*}
 &\Downarrow \\ 
 u(x) = c_1 \cos \sqrt{\lambda}x &+c_2 \sin\sqrt{\lambda}x \\
 u'(x) = -c_1\sqrt{\lambda} \sin \sqrt{\lambda}x &+c_2\sqrt{\lambda} \cos\sqrt{\lambda}x \\
 &\Downarrow u(1) - u(0)= 0, \;\;u'(1) -u'(0)=0\\
\begin{pmatrix}
 \cos \sqrt{\lambda} -1& \sin\sqrt{\lambda}\\ 
 -\sqrt{\lambda} \sin\sqrt{\lambda} & \sqrt{\lambda} \cos\sqrt{\lambda} -\sqrt{\lambda} 
\end{pmatrix} &
\begin{pmatrix}
c_1 \\ c_2
\end{pmatrix}=
\begin{pmatrix}
0 \\ 0
\end{pmatrix} \\
 &\Downarrow \\
\begin{vmatrix}
 \cos \sqrt{\lambda} -1& \sin\sqrt{\lambda}\\ 
 -\sqrt{\lambda} \sin\sqrt{\lambda} & \sqrt{\lambda} \cos\sqrt{\lambda} -\sqrt{\lambda} 
\end{vmatrix} &= 0\\ 
 &\Downarrow \\
 2\sqrt{\lambda} (1-\cos\sqrt{\lambda}) &=0 \\
 &\Downarrow \\
 \lambda &= (2n\pi)^2, \;\; n =0,1,2,\cdots \\
 u_n(x) &= \cos2n\pi x, \; \sin 2n\pi x 
\end{align*}$

**2.** $\text{ }$ Solve

>$\begin{align*}
 u_t &= u_{xx} -au, \;\; 0<x<\pi,\;\; t>0 \\ 
 u(0,t)&= 0\\ 
 u_x(\pi,t)&= 0\\ 
 u(x,0)&= x(\pi-x)
\end{align*}$

**solution** $\text{ }$ 

> $\begin{align*}
 &\Downarrow \;\; u(x,t)=X(x)T(t)\\
 \\
 \frac{T'}{T} +a &= \frac{X''}{X} = -\lambda < 0 \\
 \\
 &\Downarrow \\
 \\
 X''&+\lambda X=0, \;\;X(0)=X'(\pi)=0 \\
 T' &+(\lambda +a)T = 0 \\
 \\
 &\Downarrow \\
 \\ 
 \lambda_n &=\left( \frac{2n-1}{2} \right)^2, \;\;n =1,2,3,\cdots\\
 X_n &= \sin\frac{2n-1}{2}x \\
  T_n &= \exp\left[ -\left( \left( \frac{2n-1}{2}\right)^2 +a\right)
t\right] \\
 \\
 &\Downarrow \\
 \\
 u(x,t) &=\sum_{n=1}^\infty c_n \exp\left[ -\left( \left( \frac{2n-1}{2}\right)^2 +a\right)t\right] \sin\frac{2n-1}{2}x \\
 \\
 &\Downarrow t=0\\
 \\
x(\pi -x)&=\sum_{n=1}^\infty c_n \sin\frac{2n-1}{2}x \\
 \\
 &\Downarrow\\
 \\
 \int_0^\pi x(\pi -x) \sin \frac{2n -1}{2}x \, dx &= c_n \int_0^\pi \sin^2 \frac{2n -1}{2}x \,dx \\
 c_n &= 8 \left[ \frac{4}{\pi} (2n -1)^{-3} +(-1)^n (2n -1)^{-2} \right]
 \end{align*}$

**3.** $\text{ }$ Find a series solution for the problem

>$\begin{align*}
 u_{tt} -c^2 \left[u_{rr} +\frac{1}{r} u_r +\frac{1}{r^2}u_{\theta\theta} +u_{zz} \right ]&= 0, \;\; r< 1, \;\; 0<z<L,\;\;t>0 \\ 
 u(1,\theta,z,t)&=0 \\ 
 u(r,\theta,0,t)&=0 \\ 
 u_z(r,\theta,L,t)&=0  \\ 
 u(r,\theta,z,0)&=f(r,\theta,z) \\ 
 u_t(r,\theta,z,0)&= 0
\end{align*}$

> This problem represents the acoustic motion of a cylinder (for example, an organ pipe) fixed at one end and free at the other

**solution**

>$\begin{align*}
 &\Downarrow \;\;  u(r,\theta,z,t) = R(r) \Theta(\theta) Z(z) T(t) \\
 \\
 r^2 &\left [ \frac{T''}{c^2 T} -\frac{R'' +\frac{1}{r}R'}{R} -\frac{Z''}{Z}\right ] = \frac{\Theta''}{\Theta} = -\mu < 0 \\
 \frac{T''}{c^2 T} &-\frac{R'' +\frac{1}{r}R'}{R} +\frac{\mu}{r^2}=\frac{Z''}{Z} = -\nu < 0 \\
 \frac{T''}{c^2 T} &+\nu =\frac{R'' +\frac{1}{r}R'}{R} -\frac{\mu}{r^2}= -\lambda < 0 \\
 \\
 &\Downarrow \\ 
 \\
 \Theta'' &+\mu \Theta = 0, \;\;\Theta(0)=\Theta(2\pi), \;\Theta'(0)=\Theta'(2\pi), \;\;\Rightarrow\;\; \mu_m=m^2, \;\;m=0,1,2,\cdots, \;\; \Theta_m(\theta) = \cos m\theta,\;\sin m\theta \\
 Z'' &+\nu Z = 0, \;\; Z(0)=Z'(L)=0,\;\;\Rightarrow\;\; \nu_n=\left[ \frac{(2n -1)\pi}{2L} \right]^2,\;\;n=1,2,3,\cdots, \;\;Z_n(z)=\sin \frac{(2n -1)\pi}{2L}z \\
 rR'' &+R' +\left( \lambda r - \frac{m^2}{r}\right) R = 0, \;\;R(1)=0 \;\;\Rightarrow\;\; J_m(\sqrt{\lambda_k^{(m)}})=0, \;\; k=1,2,3,\cdots, \;\; R_{mk} =J_m(\sqrt{\lambda_k^{(m)}}r) \\
 T'' &+c^2 (\nu +\lambda)T = 0, \;\;T'(0)=0 \;\;\Rightarrow \;\;T_{kmn}(t)= \cos \left[ -c\sqrt{\left[ \frac{(2n -1)\pi}{2L} \right]^2 +\lambda_k^{(m)} } t \right]
 \\
 &\Downarrow \\ 
\\
u(r,\theta,z,t) &= \frac{1}{2} \sum_{k=1}^\infty \sum_{n=1}^\infty c_{k0n} J_0(\sqrt{\lambda_k^{(0)}}r) \sin \frac{(2n -1)\pi}{2L}z
\cos \left[ -c\sqrt{\left[ \frac{(2n -1)\pi}{2L} \right]^2 +\lambda_k^{(0)} } t \right] \\
&+\sum_{m=1}^\infty \sum_{k=1}^\infty \sum_{n=1}^\infty \left(c_{kmn}\cos m\theta +d_{kmn} \sin m\theta \right) J_m(\sqrt{\lambda_k^{(m)}}r) \sin \frac{(2n -1)\pi}{2L}z
\cos \left[ -c\sqrt{\left[ \frac{(2n -1)\pi}{2L} \right]^2 +\lambda_k^{(m)} } t \right]
\\
\\
 &\Downarrow\;\;t=0 \\ 
 \\
 f(r,\theta,z) &= \frac{1}{2} \sum_{k=1}^\infty \sum_{n=1}^\infty c_{k0n} J_0(\sqrt{\lambda_k^{(0)}}r) \sin \frac{(2n -1)\pi}{2L}z
\\
&+\sum_{m=1}^\infty \sum_{k=1}^\infty \sum_{n=1}^\infty \left(c_{kmn}\cos m\theta +d_{kmn} \sin m\theta \right) J_m(\sqrt{\lambda_k^{(m)}}r) \sin \frac{(2n -1)\pi}{2L}z \\
\\
 &\Downarrow \\ 
 \\
\frac{J^2_{m+1}\left(\sqrt{\lambda_k^{(m)}}\right)}{2} &\cdot \pi \cdot \frac{L}{2} \cdot c_{kmn} = \int_0^1 \int_0^{2\pi} \int_0^L f(r,\theta,z)\, J_m\left( \sqrt{\lambda_k^{(m)}} r\right)\,\cos m\theta \,\sin \frac{(2n -1)\pi}{2L}z \,r \,dz \, d\theta \, dr \\
\frac{J^2_{m+1}\left(\sqrt{\lambda_k^{(m)}}\right)}{2} &\cdot \pi \cdot \frac{L}{2} \cdot d_{kmn} = \int_0^1 \int_0^{2\pi} \int_0^L f(r,\theta,z)\, J_m\left( \sqrt{\lambda_k^{(m)}} r\right)\,\sin m\theta \,\sin \frac{(2n -1)\pi}{2L}z \,r \,dz \, d\theta \, dr 
\end{align*}$

**4.** $\text{ }$ You want to solve

>$\begin{align*}
 \nabla^2 u &= 1, \;\;\;\;\;\;\, r < 1 \\ 
 u(1,\theta) &= \sin\theta, \;\; 0 \leq \theta \leq 2\pi 
\end{align*}$

> Start by trying to find any particular solution of $\nabla^2 u =1$ by substituting $u_p(r,\theta)=Ar^2$ into the differential equation. After finding a particular solution, consider letting $u = w+u_p$


**solution**


>$\begin{align*}
    \nabla^2 u &=u_{rr} +\frac{1}{r} u_r +\frac{1}{r^2}u_{\theta\theta} =1\\ \\
 &\Downarrow \;\;  \text{particular solution }\; u_p = Ar^2 \\
 \\
 2A +\frac{1}{r} 2Ar &+\frac{1}{r^2} \cdot 0 = 1 \;\;\Rightarrow \;\; A=\frac{1}{4}\\
 \\
  &\Downarrow \;\;  u = w + \frac{1}{4} r^2 \\ \\
 \\
  \nabla^2 w&= 0, \;\;\;\;\;\;\, r < 1 \\ 
 w(1,\theta) &= -\frac{1}{4} +\sin\theta, \;\; 0 \leq \theta \leq 2\pi \\
 \\
   &\Downarrow \;\;  w = R(r)\Theta(\theta) \\
 \\
  - \frac{r^2R''+rR'}{R} &+\frac{\Theta''}{\Theta} = -\lambda < 0\\ 
  \\ &\Downarrow \\ \\
 \Theta'' +\lambda \Theta = 0,& \;\; \Theta(0)=\Theta(2\pi), \;\;\Theta'(0)=\Theta'(2\pi) \;\;\Rightarrow \;\; \lambda_n = n^2, \;\;n=0,1,2,\cdots, \;\;\Theta_n(\theta) =\cos n\theta, \;\sin n\theta \\
 r^2 R_n'' +r R_n'&-n^2R_n = 0, \;\; R_n(r) = r^n\; (\because u \text{ is bounded}) \\ \\
 &\Downarrow \\ \\
 w(r,\theta) = \frac{1}{2} a_{0} &+\sum_{n=1}^\infty r^n \left( a_{n} \cos n\theta +b_{n} \sin n\theta \right) \\ \\
   &\Downarrow \;\; r = 1\\ \\
-\frac{1}{4} +\sin\theta = \frac{1}{2} a_0 &+\sum_{n=1}^\infty \left( a_{n} \cos n\theta +b_{n} \sin n\theta \right) 
\;\;\Rightarrow \;\; a_0=-\frac{1}{2}, \;\;b_1=1 \\ \\
   &\Downarrow \\ \\
 w(r,\theta) =-\frac{1}{4} +r\sin\theta \;\;&\Rightarrow \;\; u(r,\theta)=-\frac{1}{4} +r\sin\theta+\frac{1}{4} r^2
\end{align*}$
